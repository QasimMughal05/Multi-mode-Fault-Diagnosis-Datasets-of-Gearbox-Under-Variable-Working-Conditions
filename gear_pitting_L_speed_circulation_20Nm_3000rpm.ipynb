```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/gear_pitting_L_speed_circulation_20Nm-3000rpm(in).csv'
data = pd.read_csv(file_path)

# Check for missing values
if data.isnull().sum().any():
    print("Missing values found! Cleaning the data...")
    data = data.dropna()
else:
    print("No missing values found.")

# Preview the dataset
print("Preview of the dataset:")
print(data.head())

Q1_speed = data['speed'].quantile(0.25)
Q3_speed = data['speed'].quantile(0.75)
IQR_speed = Q3_speed - Q1_speed
lower_bound_speed = Q1_speed - 1.5 * IQR_speed
upper_bound_speed = Q3_speed + 1.5 * IQR_speed

Q1_torque = data['torque'].quantile(0.25)
Q3_torque = data['torque'].quantile(0.75)
IQR_torque = Q3_torque - Q1_torque
lower_bound_torque = Q1_torque - 1.5 * IQR_torque
upper_bound_torque = Q3_torque + 1.5 * IQR_torque

# Filter out rows with outliers
filtered_data = data[(data['speed'] >= lower_bound_speed) & (data['speed'] <= upper_bound_speed)]
filtered_data = filtered_data[(filtered_data['torque'] >= lower_bound_torque) & (filtered_data['torque'] <= upper_bound_torque)]

# Preview the cleaned data
print("Preview of the cleaned data:")
print(filtered_data.head())

# Plot speed and torque on the same graph
plt.figure(figsize=(10, 6))
plt.plot(filtered_data.index, filtered_data['speed'], label='Speed', color='blue')
plt.plot(filtered_data.index, filtered_data['torque'], label='Torque', color='orange')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Speed and Torque vs Time')
plt.legend()
plt.grid(True)
plt.show()

# Brief description of the process
description = """
1. Loaded the dataset and checked for missing values.
2. Cleaned the data by removing rows with missing values.
3. Applied outlier removal using the IQR method for the 'speed' and 'torque' columns.
4. Plotted the speed and torque values against time on the same graph.
"""

print(description)
```
